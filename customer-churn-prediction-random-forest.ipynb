2. randomized cv: selecting random parameter

* n_estimator: number of tree in a forest

* max_depth: the maximu depth of each decision tree

* min_sample_split: min no. of sample required to split a node

* min_sample_leaf: min no. of sample required to be at leaf node

* max_feature: no. of features to consider while looking at best split

* Bootstrap: ture, false

1. Bootstrap is a method where multiple models are trainged on different random subsets of the training data using bootstrap == sampling with replacement

### ROC and AUC

* it will give probability

* how do we know 0.3 means no and 0.9 means yes -> that is where roc and auc comes


ROC - Reciver operating characteristics: is test of different the threshold value

* A graph that show how well the model seperate yes or no across all the threshold vlaues

* It draws a graph between true positive rate and false positive rate

x axis - false positive rate

y axis - true positive rate: (sensitivity) out of all actual poisitive, how many did the model correctly indentify as positive

* specificity: true negative rate: out of all actual negative, how many did the model correctly identify as negative

* roc tells, if I change my cutoof, how good is my model at catching positive value without catching too many false values
